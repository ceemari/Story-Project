# Program Planning

## The First Step: Understand the Problem

When you're learning programming, reading through a problem once is **not enough** to understand it.

Understanding a problem means that you recognize what it's asking of you, and can visualize what a successful solution looks like.

The best way to do this is to re-read the problem multiple times, then ask yourself _as many questions_ about the problem as you can. Doing this will help you explore your understanding of the problem, and better generate a list of tasks to execute.

You should always ask yourself these questions:

* What will the program look like when it runs successfully?
* What parts of the problem could cause errors or other pitfalls during development?
* What part of the problem will be the hardest to solve?
* What part of the problem will be the easiest to solve?
* Do any concepts or "building blocks" stand out as being critical to solve this problem?

Additionally, for this project, you should ask yourself the following:

* How many possible paths can this story take?
* How do I isolate a single story that the user decides to select, and not show anything from the other stories?

## The Second Step: Create the Algorithm

An algorithm is a set of steps; a list of **instructions** for the computer to follow. It specifies what you need to _do_ to solve a problem, not _how_ each step gets done.

The goal of this step is to look at the problem and generate "instructions for the computer", that we will ultimately **translate into Python**.

<font color="red">During this process, we **do not** think about the Python required to resolve a given step/task, simply **what the next step is in solving the problem**.</font>

A good way to look at this is to imagine that your computer can speak your native language, and we don't have to use Python or any programming to tell it what to do. What instructionswould you give it?

For example, if we were making a Peanut Butter and Jelly sandwich, the algorithm to achieve this goal would be made up of the following steps:

* Grab two slices of bread
* Spread Jelly on one slice
* Spead Peanut Butter on the other
* Put the two pieces together so that their condiments touch

The "how" - which fingers and muscles to move, what direction to spread the knife, the amount of pressure required - isn't important at this step. That's the final step. For now, we need a _path_, a direction to guide our hand while we code, and that is made up of the steps!

## The Third Step: Putting Pen to Paper

Once you've thought about your algorithm and what approach you want to take to solving this problem, we need to _put it somewhere useful_.

Currently, you're dealing with what we call the "blank sheet of paper problem". You're staring at an empty Python file, and trying to figure out how to get started. To overcome this, we'll write our algorithm in the file as comments!

For example, if your algorithm is:

* Get a name from the user
* Count the number of letters in their name
* Tell the user how many letters they have in their name, and if their name is long or short

You would simply write this in your code editor:

```
# Get a name from the user

# Count the number of letters in their name

# Tell the user how many letters they have in their name, and if their name is long or short
```

Make sure to leave some blank space between each comment, to write your code and make it readable. Now, instead of having a "blank sheet of paper" to deal with, you're looking at an outline for your project!

## The Fourth Step: Filling in The Blanks

Now that we've laid out our steps as comments, we have order and structure to our program.

From here, we need to go through each comment and write the code that should live underneath it. But which step do we start with?

Technically, you don't need to start at the first step. Sometimes the most complicated part of the problem is the best one to start with, to get it out of the way. But your goal is to find the next step that meets two criteria:

* It's the _easiest to solve_
* It does not require previous steps to work correctly

In some cases, it may be as simple as the very last `print()` statement that displays a dummy value. Or the very first `input()` that takes in a value from the user. Whatever the case, try and find the most self-sufficient step to solve first. This provides an "anchor" to work from.

You can either move forward from this step, using it to solve the next one. But, if it is not clear what to do next, you might want to move to the _end_ of your program and **backtrack** to your anchor. After all, we know what the end product _should look like_, so it should be easier to work from there then it is to work from the first step.

Either way, fill out each comment in the order that makes the most sense to you.

<font color="orange">Make sure that for each "step" of the algorithm you test your code! That way if things go wrong, you know it's the work you're currently doing that's causing the problem, not something you wrote before.</font>


## Tips

* Test often. Test your code as soon as you've made a meaningful change
* Use [pythontutor.com](https://www.pythontutor.com) and breakpoints when you're not sure why your program is misbehaving.
